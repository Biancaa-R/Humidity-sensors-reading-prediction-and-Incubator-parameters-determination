```python
import numpy as np
import pandas as pd
humidity=pd.read_csv("humidity copy.csv",index_col="Time")
```

```python
list1=[]
x=""
for i in list(humidity["RelativeHumidity"]):
    if i >=20 and i<32:
        x="20-32"
    elif i>32 and i<55:
        x="32-55"
    elif i>55 and i<80:
        x="55-80"
    elif i>80 and i<95:
        x="80-95"
    list1.append(x)
humidity["range"]=list1
humidity
```

```python
humidity.to_csv("range.csv")
```
