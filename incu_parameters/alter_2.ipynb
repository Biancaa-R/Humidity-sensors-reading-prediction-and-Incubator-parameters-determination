```python
import numpy as np
import pandas as pd
humidity=pd.read_csv("refined.csv")
humidity.sort_values(by=["weight"],ascending=True,inplace=True)
humidity.head()
```

```python
import pandas as pd
rates=pd.read_csv("New data set.csv")
rates.columns=rates.columns.str.lower()
rates.sort_values(by=["weight"],ascending=True)
rates.head()
```

```python
weight=list(rates["weight"])
volume=list(rates["volume_ml"])
resp=list(rates["respiration_rate"])
weight_1=[]
for i in weight:
    i=int(i*1000)
    weight_1.append(i)

print(weight_1)
weight_1.sort()
print(weight_1)
```

```python
import math
volume_arr=[]
resp_arr=[]
weight_x=list(humidity["weight"])
print(weight_x)
for i in weight_x:
    for j in weight_1:
        if i<= weight_1[0]:
            #Sambavam
            volume_arr.append(volume[0])
            resp_arr.append(resp[0])

        x=i
        if i==j:
            #Sambavam
            volume_arr.append(volume[weight_1.index(j)])
            resp_arr.append(resp[weight_1.index(j)])
            continue
        else:
            y=weight_1.index(j)
            x_=abs(i-weight_1[y-1])
            x= abs(i-j)
            if x_<x:
                #Therefore x_ was the lowest difference
                #Sambavam
                volume_arr.append(volume[weight_1.index(j)-1])
                resp_arr.append(resp[weight_1.index(j)-1])
                continue
resp_arr=resp_arr[:96]
volume_arr=volume_arr[:96]
humidity["resp_rate"]=resp_arr
humidity["volume"]=volume_arr 
```
