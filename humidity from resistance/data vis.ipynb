Plotting graph from preexisting data:

```python
import csv
import matplotlib.pyplot as plt
F=open("humidity.csv","r",newline="")
r=csv.reader(F,delimiter="|")
#so we know the order is time,rh,temp,resistance
x=1
humidity=[]
resistance=[]
for val in r:
    x=x+1
    if x>3:
        #print(val[0])
        list1=val[0].split(",")
        humidity.append(float(list1[1]))
        resistance.append(float(list1[-1]))
print(humidity)
print(resistance)

plt.scatter(resistance,humidity)
plt.xlabel("resistance")
plt.ylabel("relative humidity")
plt.show()

humidity.sort()
resistance.sort()

plt.plot(resistance,humidity)
plt.xlabel("resistance")
plt.ylabel("relative humidity")
plt.show()
```
